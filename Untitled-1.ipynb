```python
import yfinance as yf
df = yf.download("NDX" , start="2020-01-01")
print(df.head())
```

```python
import matplotlib.pyplot as plt
plt.plot(df["Close"])
plt.xlabel("Years")
plt.ylabel("Price")
plt.title("NDX price since 2020")
plt.show()
```

```python
df = yf.download("NDX", period="7y")
df["MA"] = df[("Close" ,"NDX")].rolling(200).mean()
df
```

```python
plt.plot(df["MA"])
plt.plot(df[("Close","NDX")])
plt.title("Moving average vs NDX")
plt.xlabel("Years")
plt.ylabel("Price")
plt.show()
```
